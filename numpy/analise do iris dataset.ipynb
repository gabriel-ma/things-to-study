```python
import numpy as np
amostras_iris = np.genfromtxt('iris.data.txt', delimiter=',', usecols=(0,1,2,3))
```

```python
amostras_iris
```

```python
#primeira coluna
amostras_iris[:,0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#iris setosa
plt.plot(amostras_iris[:50,0], marker='s', ms=8)
```

```python
plt.plot(amostras_iris[50:100,0], marker='s', ms=8)
```

```python
#Comprimento das sepalas, isso mostra que a iris setosa tem a menor e a virginica a maior
plt.plot(amostras_iris[:50,0], c='red', marker='s', ms=8, label='Iris Setosa')
plt.plot(amostras_iris[50:100,0], c='purple', marker='o', ms=8, label='Iris Versicolor')
plt.plot(amostras_iris[100:150,0], c='yellow', marker='v', ms=8, label='Iris Virginica')
plt.legend()
plt.show()
```

```python
#largura as sepalas, normalmente, as da setosa são maiores
plt.plot(amostras_iris[:50,1], c='red', marker='s', ms=8, label='Iris Setosa')
plt.plot(amostras_iris[50:100,1], c='purple', marker='o', ms=8, label='Iris Versicolor')
plt.plot(amostras_iris[100:150,1], c='yellow', marker='v', ms=8, label='Iris Virginica')
plt.legend()
plt.show()
```
