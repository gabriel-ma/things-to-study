```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crimes_rate = pd.read_csv('reported.csv')
```

```python
crimes_rate.head()
```

```python
crimes_rate.tail()
```

```python
plt.plot(crimes_rate['Year'], crimes_rate['crimes.total'], '-r')
plt.plot(crimes_rate['Year'], crimes_rate['crimes.person'], '-b')
```

```python
fig, ax = plt.subplots()
ax.plot(crimes_rate['Year'], crimes_rate['crimes.total'], '-r')
ax.plot(crimes_rate['Year'], crimes_rate['crimes.person'], '-b')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(crimes_rate['Year'], crimes_rate['crimes.total'], '-r')
ax.plot(crimes_rate['Year'], crimes_rate['crimes.person'], '-b')
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Crimes')
ax.set_title('Crimes rates X person')
```
