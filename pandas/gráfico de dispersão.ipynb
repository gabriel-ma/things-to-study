```python
iris = pydataset.data('iris')
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'])
```

```python
iris['Species'].unique()
```

```python
def species_number(species):
    if species == 'setosa':
        return 0
    elif species == 'versicolor':
        return 1
    else:
        return 2
```

```python
iris['Species'] = iris['Species'].apply(species_number)
```

```python
iris.head()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], c=iris['Species'], cmap='viridis', alpha=0.4)
```
