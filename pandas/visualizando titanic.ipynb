```python
%matplotlib inline
import pandas as pd
import pydataset
import matplotlib.pyplot as plt
import seaborn
```

```python
titanic = pydataset.data('titanic')
```

```python
titanic['class'].value_counts().plot(kind='bar')
```

```python
titanic.groupby('survived')['class'].value_counts().plot(kind='bar')
```

```python
estados = pd.read_csv('populacao_brasil.csv')
```

```python
estados.head()
```

```python
estados['total'].plot(kind='hist')
```

```python
fig, ax = plt.subplots()
plt.hist(estados['total'],  bins=15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['porcentagem'] = estados['total'] / estados['total'].sum()
```

```python
#gráfico de pizza
plt.pie(estados['porcentagem'], labels=estados['estado'], autopct='%1.2f%%')
```
