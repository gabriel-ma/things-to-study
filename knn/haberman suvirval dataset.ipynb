```python
def distancia_euclidiana(x, y):
    x, y = np.array(x), np.array(y)
    diff = x[:-1] - y[:-1]
    dist = np.dot(diff, diff)
    return math.sqrt(dist)
```

```python
def info_dataset(amostras, verbose=True):
    if verbose:
        print('Total de amostras: %d' % len(amostras))
    rotulo1, rotulo2 = 0, 0
    for amostra in amostras:
        if amostra[-1] == 1:
            rotulo1 += 1
        else:
            rotulo2 += 1
    if verbose:
        print('Total rotulo 1: %d' % rotulo1)
        print('Total rotulo 2: %d' % rotulo2)
    return [len(amostras), rotulo1, rotulo2]
_, rotulo1, rotulo2 = info_dataset(patients.values)
p = 0.6
treinamento, teste = [], []
max_rotulo1, max_rotulo2 = int(p * rotulo1), int(p * rotulo2)
total_rotulo1, total_rotulo2 = 0, 0
for patient in patients.values:
    if (total_rotulo1 + total_rotulo2) < (max_rotulo1 + max_rotulo2):
        treinamento.append(patient)
    else:
        teste.append(patient)
```

```python
def knn(treinamento, amostra, K):
    distancias, tamanho_treinamento = defaultdict(dict), len(treinamento)
    for i in range(tamanho_treinamento):
        distancias[i] = distancia_euclidiana(treinamento[i], amostra)
    K_vizinhos = sorted(distancias, key=distancias.get)[:K]
    
    amount_label1, amount_label2 = 0, 0
    for vizinho in K_vizinhos:
        if treinamento[vizinho][-1] == 1:
            amount_label1 += 1
        else:
            amount_label2 += 1
    if amount_label1 > amount_label2:
        return 1
    else:
        return 2
```

```python
print(treinamento[7])
print(knn(treinamento, treinamento[7], 15))
```
