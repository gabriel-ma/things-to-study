```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.target_names)
print(iris.feature_names)
```

```python
from sklearn.neural_network import MLPClassifier
x, y = iris.data, iris.target
```

```python
multi_layer_perceptron = MLPClassifier(solver='adam', learning_rate='constant', learning_rate_init=0.01,
                                        max_iter=500, random_state=1, verbose=True, momentum=0.9,
                                        hidden_layer_sizes=(5,), alpha=0.0001, activation='logistic',
                                        tol=0.0001)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
multi_layer_perceptron.fit(x_train, y_train)
outputs = multi_layer_perceptron.predict(x_test)
multi_layer_perceptron.score(x_test, y_test)
```
